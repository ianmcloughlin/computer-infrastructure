```python
# Listing files in a folder.
import os
```

```python
# List the files in the data folder.
data_files = os.listdir('../data/')

# Show.
data_files
```

## Sorting

```python
# Sort the list of files.
sorted(data_files, reverse=True)
```

```python
# Not modified.
data_files
```

```python
# Sort the list of files.
data_files.sort(reverse=True)

# Show.
data_files
```
